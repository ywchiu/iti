## 下載Excel 檔案

```python
import requests
res = requests.get('http://www.mof.gov.tw/File/Attach/72221/File_8550.xls')
with open('File_8550.xls', 'wb') as f:
    f.write(res.content)
```

## 建立函式

```python
import requests
def fileDownload(fileurl):
    res = requests.get(fileurl)
    filename = fileurl.split('/')[-1]
    with open(filename, 'wb') as f:
        f.write(res.content)
```

## 完整下載範例

```python
import requests
from bs4 import BeautifulSoup
pageurl = 'https://www.mof.gov.tw/Pages/Detail.aspx?nodeid=281&pid=72221'
res = requests.get(pageurl, verify=False)

soup = BeautifulSoup(res.text, 'html.parser')
ol = soup.select('ol ol')[0]
for a in ol.select('a'):
    if 'xls' in a['href']:
        fileurl = a['href']
        fileDownload(fileurl)
```
